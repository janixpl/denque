```python
dengue_filtered1.columns
X = dengue_filtered1[['feelslike', 'dew', 'humidity', 'precip', 'precipprob', 'precipcover', 'windspeed', 'winddir', 'sealevelpressure', 'cloudcover', 'visibility', 'uvindex', 'conditions']]
y = dengue_filtered1[['cases']]
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
train_ratio = 0.6
validation_ratio = 0.2
test_ratio = 0.2

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=1 - train_ratio)
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=test_ratio/(test_ratio + validation_ratio)) 

#print(x_train, x_val, x_test)
```

```python
import tensorflow as tf
x_train_tensor = tf.convert_to_tensor(x_train)
x_val_tensor = tf.convert_to_tensor(x_val)
x_test_tensor = tf.convert_to_tensor(x_test)
y_train_tensor = tf.convert_to_tensor(y_train)
y_val_tensor = tf.convert_to_tensor(y_val)
y_test_tensor = tf.convert_to_tensor(y_test)
```

```python
print(x_train_tensor.shape)
print(x_val_tensor.shape)
print(x_test_tensor.shape)
print(y_train_tensor.shape)
print(y_val_tensor.shape)
print(y_test_tensor.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16,input_shape=(13,), activation="relu"),
    tf.keras.layers.Dense(units=32, activation="relu"),
    tf.keras.layers.Dense(units=16, activation="relu"),
    tf.keras.layers.Dense(units=1, activation="relu")
])
```

```python
model.compile(loss="binarny_entropy", metrics=["accuracy"], optimizer='adam')
```

```python
model(x_test_tensor)
```

```python
model.predict(x_test_tensor)
```
